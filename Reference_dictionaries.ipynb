```python
myset = set(['word1', 'word2', 'word3', 'word4'])
mydict = {'word1':1, 'word5':2, 'word4':1}
mynesteddict = {'dict1':{'word1':1, 'word3':3, 'word2':5},'mydict':{'word1':1, 'word5':2, 'word4':1}}
print(myset)
print(mydict)
for word in myset:
  try:
    print(mydict[word])
  except:
    print("not present")
print("nested stuff","\n")
```

```python
print(myset)
print(mynesteddict)
for word in myset:
  for k,v in mynesteddict.items():
    try:
      print(v[word])
    except:
      print("not present")
```
