Example random matrix of numbers to do true false check

```python
from numpy.random import default_rng
rand_array = default_rng(42).random((1000,1000))
rand_array[0,1] = 0
rand_array[0,0] = 1
print(rand_array)
```

Multiplication doesn't work:

```python
print(np.matmul(rand_array, zeroarray))
print(rand_array@onearray)
```

But a comparison for boolean does from stackoverflow:

```python
x = np.array([[1,0,1],[0,0,1],[1,1,0]])
y = np.array([[1,0,1],[0,0,1],[1,1,1]])
print((x == y).sum() / x.size) #a number between 0 and 1 of similarity of two matrices in terms of number of elements on a axis
print(x == y)
```

Comparison of matrix to matrix comparison, einsum calculation and direct comparison boolean to a boolean rule.
Maybe investigate a simplification of calculation by applying a kernel function to matrix through K maps if doing several calc steps to a matrix?: https://en.wikipedia.org/wiki/Karnaugh_map

```python
import time
import tensorflow as tf
start = time.time()
print(rand_array.size)
print((rand_array == onearray).sum() / rand_array.size)
print(rand_array == onearray)
finish = time.time() - start
print("Time in seconds (compare an array == another array): \n", finish)

newstart = time.time()
e = tf.einsum('ij,jk->ik', rand_array, onearray)
print(e)
newfinish = time.time() - newstart
print("Newtime in seconds (tf.einsum): \n",newfinish)

newnewstart = time.time()
print((rand_array < 0.5).sum() / rand_array.size)
print(rand_array < 0.5)
newnewfinish = time.time() - newnewstart
print("Newnewtime in seconds(array scalars less than a value): \n",newnewfinish)

newnewstart2 = time.time()
bool_array = (rand_array < 0.5)
print(np.where(bool_array, rand_array, 0))
newnewfinish2 = time.time() - newnewstart2
print("Newnewtime2 in seconds (array scalars less than a value, then show only under a value condition): \n",newnewfinish2)

newnewstart3 = time.time()
bool_array = (rand_array < 0.5)
print(np.where(bool_array, rand_array, bool_array<0.5))
newnewfinish3 = time.time() - newnewstart3
print("Newnewtime3 in seconds (array scalars less than a value, then show only an array value under a less than condition): \n",newnewfinish3)
```

# union, intersection, difference between sets

```python
import time
a={1,2,4,5,6}
b={5,6,4,9}

start = time.time()
print(a.union(b))
fin = time.time() - start
print("union", fin)
start = time.time()
print(a.intersection(b))
fin = time.time() - start
print("intersection a from b", fin)
start = time.time()
print(b.intersection(a))
fin = time.time() - start
print("intersection b from a", fin)
start = time.time()
print(a.difference(b))
fin = time.time() - start
print("difference of a from b", fin)
start = time.time()
print(b.difference(a))
fin = time.time() - start
print("difference of b from a", fin)

start = time.time()
c=(a^b)&b
print(c)
fin = time.time() - start
print("manual diff in b", fin)
start = time.time()
c=(a^b)&a
print(c)
fin = time.time() - start
print("manual diff in a", fin)
```
